1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def contar_frecuencias(cadena):
    frecuencias = {}

    for letra in cadena:
        if letra != " ": #ignorar los espacios
            if letra in frecuencias:
                frecuencias[letra] += 1
            else:
                frecuencias[letra] = 1
    return frecuencias

texto = "Mi nombre es Joanne"
resultado = contar_frecuencias(texto)
print(resultado)
```

5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado (por defecto 5). La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota_aprobado. Si es así, el estado será "aprobado"; de lo contrario, "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
notas1 = [7,8,6,5,10]
notas2 = [4,5,4,5,4]

def evaluar_notas(lista_numeros, nota_aprobado = 5):
    media = sum(lista_numeros) / len(lista_numeros)
    
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"
    
    return (media, estado)

print(f"Resultado 1: {evaluar_notas(notas1)}") 
print(f"Resultado 1: {evaluar_notas(notas2)}") 
```

34. Crea la clase Arbol
- Define un árbol genérico con un tronco y ramas como atributos.
- Métodos disponibles: crecer_tronco, nueva_rama, crecer_ramas, quitar_rama, info_arbol.
- Código a seguir:
    - Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
    - Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
    - Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
    - Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
    - Implementar el método quitar_rama para eliminar una rama en una posición específica.
    - Implementar el método info_arbol para devolver información sobre la longitud del tronco, el número de ramas y sus longitudes.
- Caso de uso:
    - Crear un árbol.
    - Hacer crecer el tronco una unidad.
    - Añadir una nueva rama.
    - Hacer crecer todas las ramas una unidad.
    - Añadir dos nuevas ramas.
    - Retirar la rama situada en la posición 2.
    - Obtener información sobre el árbol.

```python
class Arbol:
    def __init__(self):     # 1. inicializar con tronco longitud 1 y lista vacía de ramas
        self.tronco = 1
        self.ramas = []
    
    def crecer_tronco(self):    # 2. aumentar longitud del tronco en una unidad
        self.tronco += 1
    
    def nueva_rama(self):   # 3. agregar una nueva rama de longitud 1
        self.ramas.append(1)
    
    def crecer_ramas(self):     # 4. aumentar en 1 unidad la longitud de todas las ramas existentes
        self.ramas = [longitud + 1 for longitud in self.ramas]
    
    def quitar_rama(self, posicion):    # 5- eliminar una rama en una posición específica
        if 0 <= posicion < len(self.ramas):
            self.ramas.pop(posicion)
        else:
            print("Error: Posición de rama no válida")
    
    def info_arbol(self):   # 6- devolver información sobre el árbol
        return {
            "longitud_tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitudes_ramas": self.ramas
        }

# Caso de Uso
mi_arbol = Arbol()           
mi_arbol.crecer_tronco()     # +1 = 2 troncos
mi_arbol.nueva_rama()        # 1 rama (1)
mi_arbol.crecer_ramas()      # +1 = 2 longitud rama 1 (2)
mi_arbol.nueva_rama()        # 2 ramas (2,1)
mi_arbol.nueva_rama()        # 3 ramas (2,1,1)
mi_arbol.quitar_rama(2)      # 2 ramas (2,1)

print(mi_arbol.info_arbol()) 
```

```python
class UsuarioBanco:
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def agregar_dinero(self, cantidad):
        self.saldo += cantidad

    def retirar_dinero(self, cantidad):
        if cantidad > self.saldo:
            raise ValueError(f"Fondos insuficientes de {self.nombre}")
        self.saldo -= cantidad

    def transferir_dinero(self, otro_usuario, cantidad):
        if cantidad > self.saldo:
            raise ValueError(f"Transferencia fallida: {self.nombre} no tiene saldo suficiente")
        
        self.retirar_dinero(cantidad)
        otro_usuario.agregar_dinero(cantidad)

# ------ 
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)
print(f"*** Saldos: {alicia.nombre}: {alicia.saldo} | {bob.nombre}: {bob.saldo} ***\n")

bob.agregar_dinero(20)
print(f"Bob recibe 20 - {bob.nombre}: {bob.saldo}")
print(f"*** Saldos: {alicia.nombre}: {alicia.saldo} | {bob.nombre}: {bob.saldo} ***\n")

print(f"Transferencia de 80 de {bob.nombre} a {alicia.nombre}")
try:
    bob.transferir_dinero(alicia, 80)
except ValueError as e:
    print(f"ALERTA: {e}")

print(f"*** Saldos: {alicia.nombre}: {alicia.saldo} | {bob.nombre}: {bob.saldo} ***\n")


print(f"Retirar dinero: {alicia.nombre}")
try:
    alicia.retirar_dinero(50)
    print(f"Retiro exitoso")
except ValueError as e:
    print(f"ALERTA: {e}")
print(f"*** Saldos: {alicia.nombre}: {alicia.saldo} | {bob.nombre}: {bob.saldo} ***\n")
```

36. Crea una función llamada procesar_texto
- Procesa un texto según la opción especificada: contar_palabras, reemplazar_palabras o eliminar_palabra.
- Código a seguir:
    - Crear una función contar_palabras que cuente el número de veces que aparece cada palabra en el texto y devuelva un diccionario.
    - Crear una función reemplazar_palabras para sustituir una palabra_original por una palabra_nueva en el texto y devolver el texto modificado.
    - Crear una función eliminar_palabra que elimine una palabra del texto y devuelva el texto sin ella.
    - Crear la función procesar_texto que reciba un texto, una opción ("contar", "reemplazar", "eliminar") y un número variable de argumentos según la opción elegida.
- Caso de uso:
    - Verificar el funcionamiento completo de procesar_texto.

```python
def contar_palabras(texto):
    palabras = texto.split()
    frecuencia = {}
    for p in palabras:
        p = p.lower().strip(".,")
        frecuencia[p] = frecuencia.get(p, 0) + 1
    return frecuencia

def reemplazar_palabra(texto, original, nueva):
    return texto.replace(original, nueva)

def eliminar_palabra(texto, palabra):
    return texto.replace(palabra, "").replace("  ", " ").strip()

def procesar_texto(texto, opcion, *args):
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        return reemplazar_palabra(texto, args[0], args[1])
    elif opcion == "eliminar":
        return eliminar_palabra(texto, args[0])
    else:
        return "opción no válida"

mi_texto = "Es el vecino el que elige al alcalde y es el alcalde el que quiere que sean los vecinos el alcalde"

print("Conteo:", procesar_texto(mi_texto, "contar"))
print("Reemplazo:", procesar_texto(mi_texto, "reemplazar", "alcalde", "flores"))
print("Eliminación:", procesar_texto(mi_texto, "eliminar", "que"))
```

38. Escribe un programa que determine qué calificación en texto tiene un alumno según su calificación numérica.
Reglas:
    -  0 - 69: insuficiente
    - 70 - 79: bien
    - 80 - 89: muy bien
    - 90 - 100: excelente

```python
def calificar_alumno():
    try:
        nota = float(input("Introduce la nota(0-100): "))

        if nota < 0 or nota > 100:
            print("Error: La nota debe estar entre 0 y 100")
        else:
            if nota <= 69:
                resultado = "insuficiente"
            elif nota <= 79:
                resultado = "bien"
            elif nota <= 89:
                resultado = "muy bien"
            else:
                resultado = "excelente"
            
            print(f"La calificación es: {resultado}")

    except ValueError:
        print("introduce un número válido")

calificar_alumno() # 55
calificar_alumno() # 70
calificar_alumno() # 100
calificar_alumno() # 101
calificar_alumno() # sdg
```

39. Escribe una función que tome dos parámetros: 
- figura (una cadena que puede ser "rectangulo", "circulo" o "triangulo")
- datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math

def calcular_area(figura, datos):
    figura = figura.lower()
    
    if figura == "rectangulo":
        base, altura = datos
        return base * altura
    elif figura == "circulo":
        radio = datos[0]
        return math.pi * (radio ** 2)
    elif figura == "triangulo":
        base, altura = datos
        return (base * altura) / 2
    else:
        return "figura no reconocida"

print(f"Área rectángulo: {calcular_area('rectangulo', (5, 10))}")
print(f"Área círculo: {calcular_area('circulo', (3,))}") 
print(f"Área triángulo: {calcular_area('triangulo', (4, 6))}")
```

40. Escribe un programa en Python que utilice condicionales para determinar el monto final de una compra en una tienda en línea, después de aplicar un descuento. El programa debe:
    - Solicitar al usuario el precio original de un artículo.
    - Preguntar si tiene un cupón de descuento (respuesta sí o no).
    - Si la respuesta es sí, solicitar el valor del cupón de descuento.
    - Aplicar el descuento al precio original, siempre que el valor del cupón sea válido (mayor a cero).
    - Mostrar el precio final de la compra, considerando o no el descuento.
    - Usar estructuras de control de flujo (if, elif, else) para llevar a cabo las acciones.

```python
def calcular_monto_final():
    try:
        precio_original = float(input("Introduce el precio del artículo: "))
        tiene_cupon = input("¿Tiene un cupón de descuento? (s/n): ").lower().strip()
        monto_final = precio_original

        if tiene_cupon == "s":
            valor_cupon = float(input("Introduce el descuento: "))
            
            if valor_cupon > 0:
                monto_final = precio_original - valor_cupon
                print(f"cupón válido aplicado")
            else:
                print("El cupón no es válido (debe ser mayor a cero)")
        
        elif tiene_cupon == "n":
            print("No se aplican descuentos")
        else:
            print("respuesta no reconocida")

        print(f"--- El precio final de la compra es: {monto_final:.2f}\n")

    except ValueError:
        print("Error: introduce valores numéricos")


calcular_monto_final() # 10  /si / 2 = 8
calcular_monto_final() # 10 / si /-3 -> 10
calcular_monto_final() # 10 / no -> 10
calcular_monto_final() # 10 / nsdo -> 10
```
